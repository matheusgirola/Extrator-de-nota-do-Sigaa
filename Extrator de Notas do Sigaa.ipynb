```python
linha_par= notas_sigaa.find_all("tr", "linhaPar linha")
linha_impar= notas_sigaa.find_all("tr", "linhaImpar linha")

for linha_par in linha_par:
    linha_par = linha_par.text
    linha_par = str(linha_par)
    linha_par = linha_par.replace("\t", " ")
    linha_par = linha_par.replace("\n", " ")


for linha_impar in linha_impar:
    linha_impar = linha_impar.text
    linha_impar = str(linha_impar)
    linha_impar = linha_impar.replace("\t", " ")
    linha_impar = linha_impar.replace("\n", " ")

print(linha_par, linha_impar)
```
